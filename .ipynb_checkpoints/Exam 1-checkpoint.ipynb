### The largest differences here are that each iris species has a significant difference in petal lengths. Also that the Setosa species is much much smaller than the other two.

```python
pair = sns.pairplot(
    data=df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']],
    hue="species",
    markers=["o", "s", "D"],
    corner=True
)
```

#### Some of the things I notice are that the petal length and width alone are a decent way to distingush the the different species and they are even better when used together at creating three distinct clusters. Using sepal length or sepal width with either petal length or petal width creates three nice clusters for all of those variations. Sepal width and length when compared with each other don't really yield the most helpful information because we only really get two distinct clusters.

```python
scatter = sns.lmplot(
    data=df,
    x="petal_length", 
    y="petal_width", 
    height=7,
    aspect=1.5, 
    hue="species",
).set(
    title="Petal Lengths vs Petal Widths",
)
scatter.set_axis_labels("Petal Lengths", "Petal Widths")
```

### From this we can loosely draw that the longer the length of the petal is, the larger the width of it is going to be.

# Part 3

```python
features =  [
    "petal_length",
    "petal_width",
    "sepal_width",
    "sepal_length"
]
```

```python
x = df.loc[:, features].values
```

```python
x = StandardScaler().fit_transform(x)
x
```

```python
km = KMeans(
    n_clusters = 3
)
results = km.fit(x)
```

```python
results.cluster_centers_
results.labels_
```

```python
df["cluster"] = pd.Series(results.labels_)
```

```python
df.groupby(['species', "cluster"]).size()
```

```python
def map_predictions(cluster):
    if cluster == 0:
        return "Iris-versicolor"
    elif cluster == 1:
        return "Iris-setosa"
    else:
        return "Iris-virginica"
```

```python
df["prediction"] = df["cluster"].apply(map_predictions)
```

```python
pca = PCA(n_components=2).fit_transform(x)
```

```python
pcadf = pd.DataFrame(data= pca,
                       columns=["PCA 1", "PCA 2"])
pcadf["prediction"] = df["prediction"]
pcadf["species"] = df["species"]
```

```python
scatter = sns.relplot(
    data=pcadf, 
    x="PCA 1", 
    y="PCA 2", 
    kind="scatter",
    height=7,
    aspect=1.5, 
    hue="species",
    style="prediction"
).set(
    title="Test"
)
```

##### I believe that k = 3 is correct here because there are three very distinct groups to find. The best k for a group is usually however many distinct groups there are in a set of data which sometimes isn't known which is why you have to try it out sometimes. In this case however, especially from looking at earlier graphs, it is obvious we have three groups thus that is why k=3 is the best choice. 

##### We can show that k=3 is the right choice using the principle of inertia by creating a graph of k vs inertia and finding where "the elbow" of the graph is. In this case, it would be at three because that is where we stop finding a significant decrease in inertia happening.

2. A n-dimensional Euclidean space is something that can be modeled nicely in an n-dimensional linear space. I believe the iris dataset is an example because we can model it nicely in a coordinate plane without losing any crucial data! A non-euclidean space is something like the Jacard Distance. It matters because knowing that it means we couldn't simply graph this like we did the iris dataset, we would have to do a more unique graph to display what is going on.

3. L1 is the sum of the absolute value of the difference between points and L2 is the square root of the sum of the difference between the two points squared.

4. The curse of dimensionality is the idea that each feature you add to a dataset exponentially increases the difficulty to work with it / have the computing power for it. The iris dataset does not have this problem because four features are not enough to give us very much trouble but with much more, we would have a lot of difficulties.
